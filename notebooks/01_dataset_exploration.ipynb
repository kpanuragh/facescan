# Dataset Exploration: MCD-rPPG

This notebook explores the Gaze into the Heart (MCD-rPPG) dataset structure and visualizes sample data.

```python
# Cell 1: Download dataset
from src.preprocessing.video_loader import VideoLoader
from src.config import RAW_DATA_DIR

loader = VideoLoader(cache_dir=str(RAW_DATA_DIR))
dataset = loader.download_dataset()
```

```python
# Cell 2: Explore dataset structure
print(f"Dataset size: {len(dataset)}")
print(f"Sample keys: {dataset[0].keys()}")
print(f"\nFirst sample info:")
for key in dataset[0].keys():
    val = dataset[0][key]
    print(f"  {key}: {type(val).__name__} {str(val)[:80]}...")
```

```python
# Cell 3: Load sample video
sample = dataset[0]
if 'video_path' in sample:
    frames, metadata = loader.load_video_frames(sample['video_path'])
    print(f"Loaded {metadata['loaded_frames']} frames")
    print(f"Frame shape: {frames.shape}")
    print(f"FPS: {metadata['fps']}")
else:
    print("Note: Video loading requires video_path field in dataset")
```

```python
# Cell 4: Visualize sample frames
import matplotlib.pyplot as plt
import numpy as np

if 'frames' in locals():
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))
    for i, ax in enumerate(axes):
        frame_idx = i * (len(frames) // 5)
        ax.imshow(frames[frame_idx])
        ax.set_title(f"Frame {frame_idx}")
        ax.axis('off')
    plt.tight_layout()
    plt.show()
else:
    print("Load video first to visualize frames")
```

```python
# Cell 5: Check ground truth signals
gt = loader.get_ground_truth(sample)
print("Ground truth signals available:")
for key, val in gt.items():
    if val is not None:
        if hasattr(val, 'shape'):
            print(f"  {key}: shape {val.shape}")
        else:
            print(f"  {key}: {type(val).__name__}")
    else:
        print(f"  {key}: None")
```
